# Selection sort 
- select the smallest element in the list and swap 
- repeat for all elements of the list while making the 'list' to check smaller.
-  o(n^2)
    - Check for all elements (n)
    - i+1:last search smallest (n)
    
- unstable: 
    - [3,2,3,1]-->[1,2,3,3]

```python
test = [1,4,2,1,6,3,9,6]
def SelectionSort(l):
    for i in range(len(l)):
        minpos=i
        for j in range(i+1,len(l)):
            if l[j]<l[minpos]:
                minpos=j
        l[i],l[minpos]=l[minpos],l[i]
    return l
SelectionSort(test)
```
