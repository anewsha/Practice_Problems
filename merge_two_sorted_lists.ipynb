```python
l1 = [1,5,6]
l2= [2,5,7,8]
def merge(l1,l2):
    result =[]
    i=0
    j=0
    while i<len(l1) and j<len(l2):
        if l1[i]>=l2[j]:
            result.append(l2[j])
            j+=1
        if l2[j]>=l1[i]:
            result.append(l1[i])
            i+=1
    if i<len(l1):
        result = result + l1[i+1:]
    if j<len(l2):
        result = result + l1[i+1:]
    return result
        

merge(l1,l2)
```
