 Product of Array Except Self

```python
l = [1,2,3]
```

```python
mul = []
last_i = 0
for i in range(len(l)):
    if i == last_i:
        continue
    else:
        mul = mul*l[i]
        last_i=i
mul
```

```python
l.pop(1)
```

```python
def mul_elem_in_list(l):
    mul = 1
    for i in range(len(l)):
        mul = mul*l[i]
    return mul
mul_elem_in_list(l)
```

```python
answers = []
for i in range(len(l)):         
    l_c = l.copy()
    l_c.pop(i)
    answers.append(mul_elem_in_list(l_c))
print(answers)

```

```python
suffix = {}
prefix = {}
answers = []
for i in range(len(l)):
    suffix[i]=mul_elem_in_list(l[:i])
    prefix[i]=mul_elem_in_list(l[i+1:])
    answers.append(suffix[i]*prefix[i])
answers
```
