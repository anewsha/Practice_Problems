"Write a function del_char(s, c) that takes strings s and c as input, where c has length 1 (i.e., a single character), and returns the string obtained by deleting all occurrences of c in s. If c has length other than 1, the function should return s.

Sample input-1:
- banana
- b

Output:
- anana


Sample input-2:
- banana
- an

Output:
- banana"**


```python
def del_char(s,c):
    if len(c)!=1: #not len(c)>1
        return s
    if c in s:
        string=""
        for char in s:
            if char==c:
                continue
            string+=char
        return string
    else:
        return s
start=time.perf_counter()
print(del_char("banana banana banana bananabanana banana banana ","a"))
time_v1 = time.perf_counter()-start
print("time by v1 = ", time_v1)

#------------------version2---------------------
def del_char(s,c):
    if len(c)!=1: #not len(c)>1
        return s
    l = []
    for char in s:
        if char==c:
            continue
        l.append(char)
    return "".join(l)


start=time.perf_counter()
print(del_char("banana banana banana bananabanana banana banana ","a"))
time_v2 = time.perf_counter()-start
print("time by v2 = ", time_v2)

#-----------------Version2 list comprehension---------------

def del_char(s,c):
    if len(c)!=1: #not len(c)>1
        return s
    l = [char for char in s if char!=c]
    return "".join(l)


start=time.perf_counter()
print(del_char("banana banana banana bananabanana banana banana ","a"))
time_v2_lc = time.perf_counter()-start
print("time by lc = ", time_v2_lc)

#---------------------replace method------------------

def del_char(s,c):
    if len(c)!=1: #not len(c)>1
        return s
    s = s.replace(c,"")
    return s


start=time.perf_counter()
print(del_char("banana banana banana bananabanana banana banana ","z"))
time_repl= time.perf_counter()-start
print("time by replace = ", time_repl)
```
