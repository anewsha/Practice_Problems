Write a function expanding(L) that takes a list of integer L as input and returns True if the absolute difference between each adjacent pair of elements strictly increases.

Sample Input
[1, 3, 7, 2, 9]


Output
True


Sample Input
[1, 3, 7, 2, -3]


Output
False



```python
def abs_diff(L):
    if len(L)<=2:
        return True
    last_max = abs(L[1]-L[0]) #because abs diff cant be lesser than 0
    for i in range(2,len(L)):
        this_ab_diff = abs(L[i]-L[i-1])
        if last_max>=this_ab_diff: #strictly increasing
            return False
        last_max = this_ab_diff
    return True
abs_diff([1, 3, 7, 2, -3])
```
