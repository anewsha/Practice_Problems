```python

def LinearSearch(k,l):
    for i in range(len(l)):
        if k == l[i]:
            return i
    return "not found"
k=5000
starttime= time.perf_counter()
print(LinearSearch(k,l))
time_1= time.perf_counter()-starttime
print("Time Taken = ",time_1)
```

```python
def BinarySearch(k,l):
    lidx = 0
    hidx = len(l)-1
    while lidx <= hidx:
        midx = (lidx+hidx)//2
        if k==l[midx]:
            return midx
        if k>l[midx]:
            lidx=midx
        if k<l[midx]:
            hidx=midx
    return False
k=5000
starttime= time.perf_counter()
BinarySearch(3,[1,2,3,4,5])
time_2= time.perf_counter()-starttime
print("Time Taken = ",time_2)
```

```python
def BinarySearch_Recur_noSlice(k,l,low,high):
    if low>high:
        return False
    # low=0
    # high=len(l)-1
    mid = (low+high)//2
    if l[mid]==k:
        return mid
    elif l[mid]<k:
        return BinarySearch_Recur_noSlice(k,l,low=mid+1,high=high)
    else:
        return BinarySearch_Recur_noSlice(k,l,low=low,high=mid-1)

starttime= time.perf_counter()
# l=[1,4,2,5000,5,5]
# l.sort()
high=len(l)-1
print(BinarySearch_Recur_noSlice(5000,l,0,high))
time_BinarySearch_Recur_noSlice= time.perf_counter()-starttime
print("Time Taken BinarySearch_Recur_noSlice = ",time_BinarySearch_Recur_noSlice)
```
