```python
def Find_Min_Difference(L,P):
    l = sorted(L)
    min_diff = l[4]-l[0]
    for i in range(0,len(l)-4):
        curr_diff = l[i+4]-l[i]
        # print(f"difference betw {i, i+4} which is element {l[i+4],l[i]} is {l[i+4]-l[i]}")
        if curr_diff<min_diff:
            min_diff = curr_diff
    return min_diff
Find_Min_Difference([3, 4, 1, 9, 56, 7, 9, 12],5)
```
