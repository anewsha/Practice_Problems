Given an integer n, return the sum of its digits using recursion.

Example:

123 → 6

409 → 13

```python
0//10
```

```python
def sumrecur(n):
    if n==0:
        # print("in if total,", total)
        return 0
    
    # total = total + (n%10)
    # print("while total = ",total)
    # n = n//10
    # print("while n = ",n) #12 
    sumdig = (n%10) + sumrecur(n//10)
    return sumdig
```

```python
print(sumrecur(123))
```

Given a list like:

[1, 2, 3, 4]


We want to compute the sum of the list recursively.

Same rules:

You must not use loops.

You must break the problem into a smaller version of itself.

You'll figure out the base case and recursive step.

```python
def listrecur(l):
    if not l:
        return 0
    print("l[:-1]",l[:-1])
    sum_l = l.pop() + listrecur(l)
    return sum_l
```

```python
listrecur([2,2,0,9])
```

```python
memo = {}

def change_options(amount, coins):
    if amount < 0:
        return set()
    if amount == 0:
        return {()}
    if not coins and amount > 0:
        return set()
    
    result_set = set()
    
    for i in range(len(coins)):
        if amount < coins[i]:
            continue
        new_target = amount - coins[i]
        
        if new_target in memo:
            my_temp_tuples = memo[new_target]
            print("Used memo for", new_target)
        else:
            my_temp_tuples = change_options(new_target, coins[i:])
        
        # Add current coin to each combination
        result_set.update({(coins[i], *t) for t in my_temp_tuples})
    
    memo[amount] = result_set
    print("memo[{}] = {}".format(amount, result_set))
    return result_set

print(change_options(5, [1, 2, 5, 10, 100]))
```
