```python
def maxSubArray(nums):
    maxx =0
    startidx = 0
    endidx = 0

    for i in range(len(nums)):
        # startidx = i
        temp_max=0
        for j in range(i, len(nums)):
            temp_max += nums[j]
            if temp_max> maxx:
                maxx= temp_max
                startidx = i
                endidx =j
    if startidx==endidx:
        return nums[startidx]
    return nums[startidx:endidx+1]



nums=[5,4,-1,7,8]
maxSubArray(nums)
```
